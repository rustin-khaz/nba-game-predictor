# 01 — Data Collection

Pull NBA team game logs for the 2022-23, 2023-24, and 2024-25 regular seasons.

**Output:** `data/cleaned/team_logs_all_seasons.csv`

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nba_api.stats.endpoints import leaguegamelog

print('Libraries loaded successfully.')
```

```python
# Fetch 3 seasons from the NBA API
SEASONS = ['2022-23', '2023-24', '2024-25']
all_frames = []

for season in SEASONS:
    print(f'Fetching {season}...')
    gamelog = leaguegamelog.LeagueGameLog(
        season=season,
        season_type_all_star='Regular Season',
        player_or_team_abbreviation='T',
        direction='ASC',
        sorter='DATE'
    )
    df = gamelog.get_data_frames()[0]
    all_frames.append(df)
    print(f'  Got {len(df)} rows')
    time.sleep(1)

raw = pd.concat(all_frames, ignore_index=True)
print(f'\nTotal rows fetched: {raw.shape[0]}')
raw.head(3)
```

```python
# Clean and select relevant columns
games = raw.copy()
games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])
games['WIN']       = np.where(games['WL'] == 'W', 1, 0)
games['HOME_AWAY'] = np.where(games['MATCHUP'].str.contains('vs.'), 'Home', 'Away')
games['OPPONENT']  = games['MATCHUP'].str[-3:]

games_clean = games[[
    'SEASON_ID', 'GAME_ID', 'GAME_DATE',
    'TEAM_ABBREVIATION', 'TEAM_NAME', 'OPPONENT',
    'HOME_AWAY', 'WL', 'WIN',
    'PTS', 'REB', 'AST', 'STL', 'BLK', 'TOV',
    'FG_PCT', 'FG3_PCT', 'FT_PCT', 'PLUS_MINUS'
]].copy()

games_clean = games_clean.sort_values(['TEAM_ABBREVIATION', 'GAME_DATE']).reset_index(drop=True)

print('Shape:', games_clean.shape)
print('Date range:', games_clean['GAME_DATE'].min().date(), 'to', games_clean['GAME_DATE'].max().date())
print('SEASON_ID values:', sorted(games_clean['SEASON_ID'].unique()))
games_clean.head()
```

```python
# Save to CSV
games_clean.to_csv('../data/cleaned/team_logs_all_seasons.csv', index=False)
print('Saved: data/cleaned/team_logs_all_seasons.csv')
```

```python
# 2024-25 summary — use date filter (avoids SEASON_ID type issues)
recent = games_clean[games_clean['GAME_DATE'] >= '2024-10-01'].copy()
print(f'2024-25 rows: {len(recent)}')

summary = (
    recent.groupby('TEAM_NAME')
    .agg(
        games_played   = ('GAME_ID',       'count'),
        wins           = ('WIN',           'sum'),
        avg_pts        = ('PTS',           'mean'),
        avg_tov        = ('TOV',           'mean'),
        avg_fg_pct     = ('FG_PCT',        'mean'),
        avg_plus_minus = ('PLUS_MINUS',    'mean')
    )
    .assign(win_pct=lambda x: x['wins'] / x['games_played'])
    .sort_values('win_pct', ascending=False)
    .reset_index()
)

print(f'Teams: {len(summary)}')
summary.head(10)
```

```python
# Bar chart — win percentage by team
plot_data = summary.sort_values('win_pct', ascending=True)

fig, ax = plt.subplots(figsize=(10, 9))
ax.barh(
    plot_data['TEAM_NAME'],
    plot_data['win_pct'],
    color=['steelblue' if w >= 0.5 else 'lightcoral' for w in plot_data['win_pct']]
)
ax.axvline(0.5, color='black', linestyle='--', linewidth=1, label='.500')
ax.set_xlabel('Win Percentage')
ax.set_title('NBA Team Win Percentage — 2024-25 Regular Season')
ax.legend()
plt.tight_layout()
plt.savefig('../visuals/win_percentage.png', dpi=150)
plt.show()
print('Chart saved to visuals/win_percentage.png')
```
